```python
import cv2
import matplotlib.pyplot as plt

img=cv2.imread(r"C:\Users\korkm\Downloads\11.2 smile.jpg.jpg",1)#BGR
# resmi bgr şeklinde okuyum plt ile rgb formatında gösterdik bu yüzden farklı çıkıyor
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

img_g=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img)
plt.imshow(img_g,cmap="gray",interpolation="BICUBIC")# resmin renk haritasının gray olduğunu belirttik
plt.show()

```

```python
# IMAGE STATISTIC   
import cv2 
import matplotlib.pyplot as plt 
import numpy as np
img=plt.imread(r"C:\Users\korkm\Downloads\11.2 smile.jpg.jpg")
print("min value:",img.min())# sıfır siyah demek
print("max value:",img.max())
print("mean value:",img.mean())
print("median:",np.median(img))
print("average:",np.average(img))
print("mean1: ",np.mean(img))
plt.imshow(img)
```
