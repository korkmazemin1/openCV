```python
# daha işe yarar bir fonksiyon 2. kısımda aşağı in

import cv2
import numpy as np

ykl=cv2.VideoCapture(r"C:\Users\korkm\Downloads\5.1 car.mp4.mp4")
# arka plan yakalamada temek mantık videodaki ilk kare ile diğer karelerin karşılaştırılıp ortak ve farklı görüntülerin saptanmasıdır
_,ilk_kare=ykl.read()
#arka plan çıkarm uygulamalarınd renk tonunu gri tonuna çevirip yumuşatmak(blur) işimizi kolaylaştıracak
ilk_gri=cv2.cvtColor(ilk_kare,cv2.COLOR_BGR2GRAY)
ilk_gri=cv2.GaussianBlur(ilk_gri,(5,5),0)

while 1:
    _,kare=ykl.read()
    gri=cv2.cvtColor(kare,cv2.COLOR_BGR2GRAY)
    
    diff= cv2.absdiff(ilk_gri,gri)# bu fonksiyon ile karelerin karşılaştırılmasını sağladık
    _,diff=cv2.threshold(diff,50,255,cv2.THRESH_BINARY)# boz ve gri renkleri kaldırdık  

    cv2.namedWindow("kare",cv2.WINDOW_NORMAL)
    cv2.namedWindow("ilk",cv2.WINDOW_NORMAL)
    cv2.namedWindow("dif",cv2.WINDOW_NORMAL)
    
    cv2.imshow("kare",kare)
    cv2.imshow("ilk",ilk_kare)
    cv2.imshow("dif",diff)
    
    if cv2.waitKey(20) & 0xFF==ord("q"):
        break

ykl.release()
cv2.destroyAllWindows()
```
