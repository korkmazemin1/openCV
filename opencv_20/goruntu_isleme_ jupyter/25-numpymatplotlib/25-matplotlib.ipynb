```python
import matplotlib.pyplot as plt
import numpy as  np

x=np.arange(5)# 5 e kadar sayı üret
print(x)
y=x 
print(type(x))
plt.plot(x,y,"o--")# x ve y yi tablo halinde getirir#3.parametre tablo şeklini belirler#o nokta verir "o-" hem nokta hem çizgi "o--" hem nokta hem kesikli çizgi
plt.plot(x,-y)
plt.plot(-x,y,"o")
plt.title("graphic for start")
plt.show#gösterir
```

```python
import numpy as np
import matplotlib.pyplot as plt
N=10
x=np.linspace(0,10,N)# o dan 10  kadar 10 tane sayı istedik eğer int yazmasaydık virgüllüde verir
print(x)
y=x
plt.plot(x,y,"o--")
plt.axis("off")# eksenleri yok eder
plt.show()
e=[1,3,6,12,3]
m=[12,5,-98,45,0]
plt.plot(e,[ad**2 for ad in m])
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

x= np.arange(3)
print(x)
plt.plot(x,x**4)
plt.plot(x,x*25)
plt.legend(["x**4","x*25"],loc="lower left")# burada hangi verinin neye ait olduğunu belirttik ancak bunu en son plotun içine koy
# loc ile bilgilerin konumunu belirleriz
plt.grid(True)
plt.xlabel("x=burada fonksiyonun denklemini verdik")# xin adı 
plt.ylabel("y=f(x)")
plt.savefig("C:\\Users\\korkm\\Desktop\\yazilim\\goruntu_isleme_python\\.ipynb_checkpoints\\görüntü_isleme_ jupyter\\25-numpymatplotlib\\plt.png")
print(plt.axis())# maksimum ve minimum noktaların çıktılarını aldık
plt.axis([0,2,0,10])# ve y eksenlerinin sırası ile max minlerini belirledik

plt.show()
```
