```python
import cv2
import numpy as np

def bos(x):
    pass

ykl=cv2.VideoCapture(0)

cv2.namedWindow("ayarlar")

cv2.createTrackbar("Lower-Hue","ayarlar",0,180,bos)
cv2.createTrackbar("Lower-Saturation","ayarlar",0,255,bos)
cv2.createTrackbar("Lower-Value","ayarlar",0,255,bos)
cv2.createTrackbar("Upper-Hue","ayarlar",0,180,bos)
cv2.createTrackbar("Upper-Saturation","ayarlar",0,255,bos)
cv2.createTrackbar("Upper-Value","ayarlar",0,255,bos)
##burada trackbar ayarlandı
font=cv2.FONT_HERSHEY_COMPLEX

while 1:
    
    ret,kare=ykl.read()
    kare=cv2.flip(kare,1)
    
    hsv = cv2.cvtColor(kare,cv2.COLOR_BGR2HSV)



    lh=cv2.getTrackbarPos("Lower-Hue","ayarlar")
    ls=cv2.getTrackbarPos("Lower-Saturation","ayarlar")
    lv=cv2.getTrackbarPos("Lower-Value","ayarlar")
    uh=cv2.getTrackbarPos("Upper-Hue","ayarlar")
    us=cv2.getTrackbarPos("Upper-Saturation","ayarlar")
    uv=cv2.getTrackbarPos("Upper-Value","ayarlar")
    # burada trackbarlardan değerler alındı get ile
    lower_color=np.array([lh,ls,lv])
    upper_color=np.array([uh,us,uv])
    # burada ise alınan değerler dizi haline getirildi

    mask=cv2.inRange(hsv,lower_color,upper_color)

    kernel= np.ones((5,5),np.uint8)# siyah noktaları yok etmek için kullanılır

    mask = cv2.erode(mask,kernel)

    contours,_= cv2.findContours(mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

    for cnt in contours:# tespit edilen resmin merkez noktlarının eldesi için 
        alan=cv2.contourArea(cnt)
        epsilon= 0.02*cv2.arcLength(cnt,True)
        approx= cv2.approxPolyDP(cnt,epsilon,True)                             

        x = approx.ravel()[0]# approxdaki verileri tek satırda toplarız
        y= approx.ravel()[1]        


        if alan > 400:
            cv2.drawContours(kare,[approx],0,(0,0,0),5)
            if len(approx)==3:
                cv2.putText(kare,"ucgen",(x,y),font,1,(0,0,0))
            elif len(approx)==4:
                cv2.putText(kare,"dortgen",(x,y),font,1,(0,0,0))
            elif len(approx)==5:
                cv2.putText(kare,"besgen",(x,y),font,1,(0,0,0))
            elif len(approx)==6:
                cv2.putText(kare,"altigen",(x,y),font,1,(0,0,0))
            elif len(approx)==7:
                cv2.putText(kare,"yedigen",(x,y),font,1,(0,0,0))
            elif len(approx)==8:
                cv2.putText(kare,"sekizgen",(x,y),font,1,(0,0,0))
            elif len(approx)==9:
                cv2.putText(kare,"dokuzgen",(x,y),font,1,(0,0,0))
            elif len(approx)==10:
                cv2.putText(kare,"ongen",(x,y),font,1,(0,0,0))
            elif len(approx)>10:
                cv2.putText(kare,"cember",(x,y),font,1,(0,0,0))


    cv2.imshow("yakalama",kare)
    cv2.imshow("maske",mask)

    if cv2.waitKey(3)& 0xFF ==ord("q"):
        break

ykl.release()    
cv2.destroyAllWindows()
```
